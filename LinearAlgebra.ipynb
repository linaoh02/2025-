2. **벡터, 행렬 연산, 그래프 그리기**

```python
import numpy as np
```

```python
def print_val(x) :
  print ("Type: ", type(x))
  print ("Shape: ", x.shape)
  print ("값:\n", x)
  print (" ")
```

```python
x = np.array([1,2,3]) #넘 파일에 어레이로
print_val(x)

x[0] = 5
print_val(x) #시험에서 빈칸 뚫어서 쓰게하기 = np.array([1,2,3]) #넘 파일에 어레이로

x[0] = 5
print_val(x) #시험에서 빈칸 뚫어서 쓰게하기
```

```python
y = np.array([[1,2,3],[4,5,6]])
print_val(y)

a = np.ones((3,2)) #보통 셰입을 나타낼때 (3행 2열)(행: 가로줄 열:세로줄)
print_val(a)

a = np.zeros((2,2))
print_val(a)

a = np.eye(3,3) #아이니까...eye.....
print_val(a)
```

```python
a = np.random.random((4,4)) #4,4짜리 행렬에 랜덤한 값이 들어감
print_val(a)

a = np.random.randn(4,4) #rand'n'<- normal(가우시안) distribution / 0에 가까운 값들이 나올 확률이 높음
print_val(a)
```

```python
b = a[:2, 1:3] #행0~1, 열 1~2
print_val(b)

row1 = a[1, :] #1번째 행
print_val(row1)
```

```python
m1 = np.array([[1,2], [3,4]], dtype = np.float64)
m2 = np.array([[5,6], [7,8]], dtype = np.float64)

#elementwise 원소별로 sum더하라
print_val(m1+m2)
print_val(np.add(m1, m2))
```

```python
#elementwise difference
print_val (m1-m2)
print_val(np.subtract(m1,m2))

#elementwise product
print_val(m1*m2)
print_val(np.multiply(m1,m2))
```

```python
#elementwise division
print_val(m1/m2)
print_val(np.divide(m1,m2))

#elementwise square root
print_val(np.sqrt(m1)) #sqrt = 스퀘어 루트
```

```python
m1 = np.array([[1,2],[3,4]]) #(2,2)
m2 = np.array([[5,6],[7,8]]) #(2,2)

v1 = np.array([9,10]) #(2,)
v2 = np.array([11,12]) #(2,)

print_val(m1)
print_val(m2)
print_val(v1)
print_val(v2)
```

벡터-벡터 연산

```python
print_val(v1.dot(v2)) #dot이란 dot product의 줄임말
print_val(np.dot(v1,v2))
```

벡터-행렬 연산

```python
print_val(m1.dot(m1)) #(2,2) * (2,1) -> (2,1)

print_val(np.dot(m1,v2))
# m1 = (1  2)  v1 = ( 9)   m1*v1= (9+20 = 29) = (29)
#      (3  4)       (10)          (27+40 = 67)  (67)
```

```python
#전치 행렬
print_val(m1)
print_val(m1.T)
```

```python
print_val(np.sum(m1)) #행렬의 모든 원소의 합
print_val(np.sum(m1, axis=0)) #shape[0](행)을 압축시키자
print_val(np.sum(m1, axis=1)) #shape[1](열)을 압축시키자
```
